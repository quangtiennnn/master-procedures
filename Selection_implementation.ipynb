```python
%%capture
# from dataframe_selection import *
from dataframe_selection.Selection import Selection
from dataframe_selection.tools import *
import pandas as pd
```

# Import Data

```python
df = pd.read_csv('database/seaside_hotel_reviews.csv', encoding='utf-8-sig')
```

```python
sample_size = 100  
sample = df.sample(n=sample_size, random_state=42)
```

# Selection

```python
selection = Selection(sample)
```

```python
selection.hotel
```

```python
selection.hotel_review
```

# Filtering

```python
['province','region','star_rating','coordinate','region','dis2coast','seaside']
```

```python
selection.province
```

```python
filtered_hotel = selection.filtering(
    province=['Da Nang'],
    star_rating=[3],
    # region=['Central Vietnam'],
    # dis2coast=[0, 0.5],
    # seaside=1
)
filtered_hotel
```

```python
filtered_review = selection.filtering(
    extracting = 'review',
    province=['Da Nang'],
    star_rating=[4],
    region=['Central Vietnam'],
    dis2coast=[0, 1],
    seaside=1
)
filtered_review
```

```python
filtered_review.value_counts('region')
```
