```python
import pandas as pd
import numpy as np
import pycountry

# Function to map country names to their ISO 3166-1 alpha-2 codes
def country_to_iso_code(country):
    if pd.isna(country):
        return 'unknown'
    try:
        # Lookup country code using pycountry
        return pycountry.countries.lookup(country).alpha_2
    except LookupError:
        return 'unknown'

# Example array
countries = df.national.unique()

# Convert the array to a pandas DataFrame
df = pd.DataFrame(countries, columns=['country'])

# Apply the function to the DataFrame
df['country_code'] = df['country'].apply(country_to_iso_code)

# Print the DataFrame
print(df)
```
